```python
## Single mean hypothesis testing
```

```python
import pymc3 as pm
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
%matplotlib inline
```

### Unknown mean with known variance
#### Model specification: 

prior on $\mu$
$$\mu \sim N(0,1000)$$ 
Likelihood of the data given the parameter
$$y \sim N(\mu,1)$$ 

```python
#Generating some random data to represent observed evidence
y = np.random.normal(0.5,1,1000)
```

```python
y
```

```python
# Code to estimate mu given the data values y
n_samples = 500
with pm.Model() as model:
    mu = pm.Normal('mu', mu = 0, sd = 10) # declaring the prior on mu
    likelihood = pm.Normal('y',mu=mu,sd=1,observed = y) #defining the likelihood function
    trace = pm.sample(n_samples)#, njobs=4)
```

```python
#Posterior distribution plots
pm.traceplot(trace)
```

### Unknown mean and unknown variance
#### Model specification: 

prior on $\sigma$
$$\sigma \sim exp(\lambda) $$

prior on $\mu$
$$\mu \sim N(0,\sigma)$$ 

Likelihood of the data given the parameter
$$y \sim N(\mu,\sigma)$$ 

```python
# Code to estimate mu given the data values y
n_samples = 500
with pm.Model() as model:
    sigma = pm.Exponential('sigma',lam = 1/5) # prior on sigma
    mu = pm.Normal('mu', mu = 0, sd = sigma ) # declaring the prior on mu   
    likelihood = pm.Normal('y',mu=mu,sd=sigma,observed = y) #defining the likelihood function
    trace1 = pm.sample(n_samples)#, njobs=4)
```

```python
pm.traceplot(trace1)
```
